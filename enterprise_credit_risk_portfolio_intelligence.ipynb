```python
import pandas as pd
import numpy as np

# =====================================================
# ENTERPRISE CREDIT RISK & LOAN PORTFOLIO INTELLIGENCE
# =====================================================

np.random.seed(42)

# -------------------------------
# DATA GENERATION
# -------------------------------

n = 100000

customer_id = range(100001, 100001 + n)

loan_amount = np.random.randint(
    5000,
    12000,
    n
)

annual_income = np.random.randint(
    200000,
    2000000,
    n
)

credit_score = np.random.randint(
    300,
    900,
    n
)

loan_type = np.random.choice(
    ['Home Loan',
     'Personal Loan',
     'Auto Loan',
     'Business Loan',
     'Education Loan'],
    n
)

region = np.random.choice(
    ['North',
     'South',
     'East',
     'West'],
    n
)

employment_years = np.random.randint(
    0,
    30,
    n
)

# -------------------------------
# CREATE DATAFRAME
# -------------------------------

df = pd.DataFrame({
    'Customer_ID': customer_id,
    'Loan_Amount': loan_amount,
    'Annual_Income': annual_income,
    'Credit_Score': credit_score,
    'Employment_Years': employment_years,
    'Loan_Type': loan_type,
    'Region': region
})

# -------------------------------
# CREDIT RISK SCORING
# -------------------------------

conditions = [
    df['Credit_Score'] >= 750,
    (df['Credit_Score'] >= 650) &
    (df['Credit_Score'] < 750),
    df['Credit_Score'] < 650
]

risk_labels = [
    'Low Risk',
    'Medium Risk',
    'High Risk'
]

df['Risk_Category'] = np.select(
    conditions,
    risk_labels,
    default='High Risk'
)

# Debt-to-Income Ratio

df['DTI_Ratio'] = round(
    df['Loan_Amount'] /
    df['Annual_Income'],
    4
)

# Simulated Default Flag

df['Default_Flag'] = np.where(
    df['Risk_Category'] == 'High Risk',
    np.random.choice(
        [0,1],
        size=n,
        p=[0.75,0.25]
    ),
    np.random.choice(
        [0,1],
        size=n,
        p=[0.97,0.03]
    )
)

# -------------------------------
# ENTERPRISE KPIs
# -------------------------------

total_customers = len(df)

total_exposure = df['Loan_Amount'].sum()

average_loan = round(
    df['Loan_Amount'].mean(),
    2
)

default_rate = round(
    (df['Default_Flag'].mean()) * 100,
    2
)

high_risk_customers = len(
    df[
        df['Risk_Category']
        == 'High Risk'
    ]
)

# -------------------------------
# PORTFOLIO ANALYSIS
# -------------------------------

risk_exposure = (
    df.groupby('Risk_Category')
      ['Loan_Amount']
      .sum()
      .sort_values(
          ascending=False
      )
)

region_exposure = (
    df.groupby('Region')
      ['Loan_Amount']
      .sum()
      .sort_values(
          ascending=False
      )
)

loan_product_exposure = (
    df.groupby('Loan_Type')
      ['Loan_Amount']
      .sum()
      .sort_values(
          ascending=False
      )
)

# -------------------------------
# PRINT BUSINESS INSIGHTS
# -------------------------------

print("\n==============================")
print("PORTFOLIO SUMMARY")
print("==============================")

print(f"Total Customers: {total_customers:,}")
print(f"Total Loan Exposure: ₹{total_exposure:,.0f}")
print(f"Average Loan Amount: ₹{average_loan:,.2f}")
print(f"Default Rate: {default_rate}%")
print(f"High Risk Customers: {high_risk_customers:,}")

print("\n==============================")
print("RISK EXPOSURE")
print("==============================")
print(risk_exposure)

print("\n==============================")
print("REGIONAL EXPOSURE")
print("==============================")
print(region_exposure)

print("\n==============================")
print("LOAN PRODUCT EXPOSURE")
print("==============================")
print(loan_product_exposure)

# -------------------------------
# EXPORT DATASET
# -------------------------------

df.to_csv(
    'enterprise_credit_risk_dataset.csv',
    index=False
)

print("\nDataset Exported Successfully!")
```
